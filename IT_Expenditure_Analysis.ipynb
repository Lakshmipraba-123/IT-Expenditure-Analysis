```python
file_name = list(uploaded.keys())[0]
file_name
```

```python
import pandas as pd

df_actual = pd.read_excel(file_name, sheet_name=0)
df_budget = pd.read_excel(file_name, sheet_name=1)
```

```python
df_actual.head()
df_budget.head()
```

```python
df_actual['Date'] = pd.to_datetime('1-' + df_actual['Date'] + '-2000', format='%d-%b-%Y')
df_budget['Date'] = pd.to_datetime('1-' + df_budget['Date'] + '-2000', format='%d-%b-%Y')

df_actual = df_actual.fillna(0)
df_budget = df_budget.fillna(0)
```

```python
df_actual['Month'] = df_actual['Date'].dt.strftime('%Y-%m')
df_budget['Month'] = df_budget['Date'].dt.strftime('%Y-%m')
```

```python
actual_month = (
    df_actual.groupby('Month')['Actual']
    .sum()
    .reset_index()
)

budget_month = (
    df_budget.groupby('Month')['Actual']
    .sum()
    .reset_index()
)

trend = pd.merge(actual_month, budget_month, on='Month')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
plt.plot(trend['Month'], trend['Actual_x'], label='Actual Spend')
plt.plot(trend['Month'], trend['Actual_y'], label='Planned Budget')
plt.title("Actual vs Planned IT Spend (Monthly)")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
trend['Variance'] = trend['Actual_x'] - trend['Actual_y']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
sns.barplot(data=trend, x='Month', y='Variance')
plt.title("Monthly IT Spend Variance")
plt.xticks(rotation=45)
plt.show()
```

```python
df_combined = pd.merge(
    df_actual,
    df_budget,
    on=['Date','Business Area','IT Area','Country','Cost element name'],
    suffixes=('_actual', '_budget')
)

df_combined['Variance'] = (
    df_combined['Actual_actual'] - df_combined['Actual_budget']
)
```

```python
business_var = (
    df_combined.groupby('Business Area')['Variance']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=business_var, x='Variance', y='Business Area')
plt.title("IT Spend Variance by Business Area")
plt.show()
```

```python
it_area_var = (
    df_combined.groupby('IT Area')['Variance']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=it_area_var, x='Variance', y='IT Area')
plt.title("IT Spend Variance by IT Area")
plt.show()
```
